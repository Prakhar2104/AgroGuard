```python
import json
import os

# Path to your mapping file
mapping_path = r"C:\Users\tiwar\plant_disease_detection\notebook\disease_mapping.json"

# 1️⃣ Check if file exists
if not os.path.exists(mapping_path):
    print("❌ Mapping file not found!")
else:
    print("✅ Mapping file found")

    # 2️⃣ Load the JSON
    with open(mapping_path, "r") as f:
        disease_mapping = json.load(f)

    # 3️⃣ Basic checks
    print(f"✅ Loaded {len(disease_mapping)} diseases")

    # Show first 5 entries
    for i, (disease, treatment) in enumerate(disease_mapping.items()):
        print(f"{i+1}. {disease} → {treatment}")
        if i >= 4:  # show only first 5
            break

    # 4️⃣ Check for missing treatments (handles string, list, dict)
    missing = []
    for d, t in disease_mapping.items():
        if t is None:
            missing.append(d)
        elif isinstance(t, str) and t.strip() == "":
            missing.append(d)
        elif isinstance(t, (list, dict)) and len(t) == 0:
            missing.append(d)

    if missing:
        print(f"⚠️ Found {len(missing)} diseases without treatments: {missing[:5]} ...")
    else:
        print("✅ All diseases have treatments mapped!")
```
