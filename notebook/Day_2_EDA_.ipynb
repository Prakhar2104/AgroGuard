```python
import os
import matplotlib.pyplot as plt

# ‚úÖ Base path to your datasets
base_path = r"C:\Users\tiwar\plant_disease_detection\Dataset"

# ‚úÖ Paths to each dataset
dataset_paths = {
    "PlantDoc": os.path.join(base_path, "plantdoc"),
    "Agri India 20K": os.path.join(base_path, "agri_india"),
    "Wheat Disease": os.path.join(base_path, "wheat_disease")
}
```

```python
# ‚úÖ This function scans the folders and counts images
def get_class_distribution(dataset_name, dataset_path):
    print(f"\nüìÇ Dataset: {dataset_name}")
    total = 0
    class_counts = {}

    for class_name in os.listdir(dataset_path):
        class_folder = os.path.join(dataset_path, class_name)
        if os.path.isdir(class_folder):
            num_images = len(os.listdir(class_folder))
            class_counts[class_name] = num_images
            total += num_images
            print(f"üìÅ {class_name}: {num_images} images")

    print(f"üî¢ Total Images: {total}")
    return class_counts
```

```python
# ‚úÖ Store class counts for all datasets
all_distributions = {}

for dataset_name, path in dataset_paths.items():
    counts = get_class_distribution(dataset_name, path)
    all_distributions[dataset_name] = counts
```

```python
# ‚úÖ Bar chart of class distribution
def plot_distribution(dataset_name, class_counts):
    plt.figure(figsize=(14, 5))
    classes = list(class_counts.keys())
    counts = list(class_counts.values())

    plt.bar(classes, counts, color='darkgreen')
    plt.xticks(rotation=90)
    plt.title(f"{dataset_name} - Class Distribution")
    plt.xlabel("Class Names")
    plt.ylabel("Number of Images")
    plt.tight_layout()
    plt.show()
```

```python
# ‚úÖ Visualize class distribution for each dataset
for dataset, counts in all_distributions.items():
    plot_distribution(dataset, counts)
```

```python
# ‚úÖ Save summary to label_summary.txt
summary_path = r"C:\Users\tiwar\plant_disease_detection\label_summary.txt"

with open(summary_path, "w", encoding="utf-8") as file:
    for dataset, class_counts in all_distributions.items():
        file.write(f"\nüìò Dataset: {dataset}\n")
        for cls, count in class_counts.items():
            file.write(f"{cls}: {count} images\n")
        file.write("\n-----------------------\n")

print("‚úÖ Summary saved to label_summary.txt")
```

```python
import random
from PIL import Image

# ‚úÖ Function to show one random image per class
def preview_sample_images(dataset_name, dataset_path):
    print(f"\nüì∏ Previewing sample images from: {dataset_name}")
    class_names = os.listdir(dataset_path)

    for class_name in random.sample(class_names, min(5, len(class_names))):  # Show 5 random classes
        class_folder = os.path.join(dataset_path, class_name)
        if os.path.isdir(class_folder):
            image_files = os.listdir(class_folder)
            if image_files:
                img_path = os.path.join(class_folder, random.choice(image_files))
                img = Image.open(img_path)

                plt.figure(figsize=(3, 3))
                plt.imshow(img)
                plt.axis("off")
                plt.title(class_name)
                plt.show()

# ‚úÖ Show samples from each dataset
for dataset_name, dataset_path in dataset_paths.items():
    preview_sample_images(dataset_name, dataset_path)
```
